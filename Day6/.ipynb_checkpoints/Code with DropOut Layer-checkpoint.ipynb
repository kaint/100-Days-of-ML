# Architecture of our Network

### Convolution Layer1 = 32Units each with filter=5*5*1 its ouput will be                                                    28*28*32

### Max Pooling Layer1 = 2*2 ouput will be 14*14*32  
### Convolution Layer2 = 64Units with filter=5*5*32 its ouput will be 14*14*64 
### Max Pooling Layer2 = 2*2 ouput will be 7*7*64


```python
input_width=28
input_height=28
input_channel=1
input_pixel=784

n_conv1=32
conv1_k=5
stride_conv1=1
max_pool1_k=2

n_conv2=64
conv2_k=5
stride_conv2=1
max_pool2_k=2

n_hidden=1024
n_output=10
input_size_to_hidden=( input_width//( max_pool1_k*max_pool2_k ) )*( input_height//( max_pool1_k*max_pool2_k ) )*n_conv2
```

```python
#Let us first initialize the weights of our neural Network
```

```python
weights={
    'wc1':tf.Variable(tf.random_normal([conv1_k,conv1_k,input_channel,n_conv1])),   #weights for filters of convo layer 1 
    'wc2':tf.Variable(tf.random_normal([conv2_k,conv2_k,n_conv1,n_conv2])),         #weights for filters of convo layer 2
    'wh1':tf.Variable(tf.random_normal([input_size_to_hidden,n_hidden])),           #weights for filters of neural hidden1
    'wo':tf.Variable(tf.random_normal([n_hidden,n_output]))                        #weights for filters of neural hidden2
}
biases={
    'bc1':tf.Variable(tf.random_normal([n_conv1])),                                 #biases for filters of convo layer 1 
    'bc2':tf.Variable(tf.random_normal([n_conv2])),                                 #biases for filters of convo layer 2
    'bh1':tf.Variable(tf.random_normal([n_hidden])),                                 #biases for filters of neural hidden1
    'bo':tf.Variable(tf.random_normal([n_output]))                                  #biases for filters of neural hidden2
}
```

```python
def conv(x,weights,bias,strides=1):
    out=tf.nn.conv2d(x,weights,padding="SAME",strides=[1,strides,strides,1]) #input,weights,padding, strides=[no of images (1) at a time 1, height, width, no of channgels(1 along the depth no movement)]
    out=tf.nn.bias_add(out,bias)
    out=tf.nn.relu(out)
    return out 

def maxpooling(x,k=2):
    return tf.nn.max_pool(x,padding='SAME',ksize=[1,k,k,1],strides=[1,k,k,1])  #
```

```python

def cnn(x,weights,biases):
    x=tf.reshape(x,shape=[-1,input_height,input_width,input_channel])  #reshaping in matrix form to pass the filter from it
    
    conv1=conv(x,weights['wc1'],biases['bc1'],stride_conv1) #passing through  first convolution layer
    conv1_pool=maxpooling(conv1,max_pool1_k)   #passing it throught maxpooling layer
    
    conv2=conv(conv1_pool,weights['wc2'],biases['bc2'],stride_conv2)  #passing throught conv2 layer
    conv2_pool=maxpooling(conv2,max_pool2_k)             #again reducing size of image to half
    
    hidden_input=tf.reshape(conv2_pool,shape=[-1,input_size_to_hidden])
    hidden_output_before_activation=tf.add(tf.matmul(hidden_input,weights['wh1']) , biases['bh1'] )
    hidden_output=tf.nn.relu(hidden_output_before_activation)
    
    output=tf.add(tf.matmul(hidden_output,weights['wo'] ),biases['bo'] )
    
    
    return output
```

```python
x = tf.placeholder(tf.float32, [None, input_pixel])
y =tf.placeholder(tf.int32, [None, n_output])
pred = cnn(x, weights, biases)
```

```python
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=pred, labels = y))
```

```python
optimizer = tf.train.AdamOptimizer(learning_rate=0.01)
optimize = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
batch_size = 100
for i in range(25):
    print(i)
    num_batches = int(mnist.train.num_examples/batch_size)
    total_cost = 0
    for j in range(num_batches):
        print('hlo')
        batch_x, batch_y = mnist.train.next_batch(batch_size)
        c, _ = sess.run([cost,optimize], feed_dict={x:batch_x , y:batch_y})
        total_cost += c
    print('cost is '+total_cost)
```

```python
predictions = tf.argmax(pred, 1)
correct_labels = tf.argmax(y, 1)
correct_predictions = tf.equal(predictions, correct_labels)
predictions,correct_predictions  = sess.run([predictions, correct_predictions], feed_dict={x:mnist.test.images,
                                              y:mnist.test.labels})
correct_predictions.sum()
```
